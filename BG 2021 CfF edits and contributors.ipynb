```python
import urllib.request as ur
import urllib.parse
import json
import pandas as pd
import datetime as dtstring
from datetime import datetime
import numpy as np
import re
```

```python
df = pd.read_csv('WMF BG CfF 2021 talk pages list.csv')
df
```

```python
def listToString(l):
    str1 = ''
    for i in l:
        str1 += i
    return str1
```

```python
w1Start = dtstring.datetime(2021, 2, 1, 0, 0)
w1End = dtstring.datetime(2021, 2, 7, 0, 0)
w2Start = dtstring.datetime(2021, 2, 8, 0, 0)
w2End = dtstring.datetime(2021, 2, 14, 0, 0)
w3Start = dtstring.datetime(2021, 2, 15, 0, 0)
w3End = dtstring.datetime(2021, 2, 21, 0, 0)
w4Start = dtstring.datetime(2021, 2, 22, 0, 0)
w4End = dtstring.datetime(2021, 2, 28, 0, 0)
w5Start = dtstring.datetime(2021, 3, 1, 0, 0)
w5End = dtstring.datetime(2021, 3, 7, 0, 0)
w6Start = dtstring.datetime(2021, 3, 8, 0, 0)
w6End = dtstring.datetime(2021, 3, 14, 0, 0)
```

```python
for index in df.index:
    w0Edits, w1Edits, w2Edits, w3Edits, w4Edits, w5Edits, w6Edits, w7Edits = 0, 0, 0, 0, 0, 0, 0, 0
    w0Editors, w1Editors, w2Editors, w3Editors, w4Editors, w5Editors, w6Editors, w7Editors = [], [], [], [], [], [], [], []
    pageTitle = urllib.parse.quote(df.loc[index, 'title'])
    pageID = str(df.loc[index, 'pageID'])
    url = 'https://meta.wikimedia.org/w/api.php?action=query&format=json&prop=revisions&titles='+pageTitle+'&rvprop=ids%7Ctimestamp%7Cuser&rvlimit=500&rvdir=older'
    response = ur.urlopen(url)
    jsonData = json.loads(response.read())
    data1 = jsonData['query']['pages'][pageID]['revisions']
    for entry in data1:
        fullStamp = entry['timestamp']
        dateString = re.findall("\d{4}-\d{2}-\d{2}", fullStamp)
        dateString = listToString(dateString)
        dateStamp = datetime.strptime(dateString,'%Y-%m-%d')
        if dateStamp < w1Start:
            w0Edits = w0Edits+1
            w0Editors.append(entry['user'])
        elif dateStamp >= w1Start and dateStamp <= w1End:
            w1Edits = w1Edits+1
            w1Editors.append(entry['user'])
        elif dateStamp >= w2Start and dateStamp <= w2End:
            w2Edits = w2Edits+1
            w2Editors.append(entry['user'])
        elif dateStamp >= w3Start and dateStamp <= w3End:
            w3Edits = w3Edits+1
            w3Editors.append(entry['user'])
        elif dateStamp >= w4Start and dateStamp <= w4End:
            w4Edits = w4Edits+1
            w4Editors.append(entry['user'])
        elif dateStamp >= w5Start and dateStamp <= w5End:
            w5Edits = w5Edits+1
            w5Editors.append(entry['user'])
        elif dateStamp >= w6Start and dateStamp <= w6End:
            w6Edits = w6Edits+1
            w6Editors.append(entry['user'])
        else:
            w7Edits = w7Edits+1
            w7Editors.append(entry['user'])
    df.loc[index,'w0Edits'] = w0Edits
    df.loc[index,'w1Edits'] = w1Edits
    df.loc[index,'w2Edits'] = w2Edits
    df.loc[index,'w3Edits'] = w3Edits
    df.loc[index,'w4Edits'] = w4Edits
    df.loc[index,'w5Edits'] = w5Edits
    df.loc[index,'w6Edits'] = w6Edits
    df.loc[index,'w7Edits'] = w7Edits
    df.loc[index,'w0EditorsCount'] = len(list(dict.fromkeys(w0Editors)))
    df.loc[index,'w1EditorsCount'] = len(list(dict.fromkeys(w1Editors)))
    df.loc[index,'w2EditorsCount'] = len(list(dict.fromkeys(w2Editors)))
    df.loc[index,'w3EditorsCount'] = len(list(dict.fromkeys(w3Editors)))
    df.loc[index,'w4EditorsCount'] = len(list(dict.fromkeys(w4Editors)))
    df.loc[index,'w5EditorsCount'] = len(list(dict.fromkeys(w5Editors)))
    df.loc[index,'w6EditorsCount'] = len(list(dict.fromkeys(w6Editors)))
    df.loc[index,'w7EditorsCount'] = len(list(dict.fromkeys(w7Editors)))
```
